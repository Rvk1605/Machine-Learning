```python
#reading CSV File
df = pd.read_csv('https://raw.githubusercontent.com/neylsoncrepalde/projeto_eda_covid/master/covid_19_data.csv')
```

```python
df.info()
df.head() #Gives first 5 data by default
df.tail(10) #Gives last 10 Data
```

```python
df.corr() #Shows Correlation between Data
```

```python
df = df.rename(columns={'ObservationDate':'OD','Province/State':'State'})
df
```

```python
#Groupify dataset
gp = df.groupby(['State'])[['Confirmed','Deaths','Recovered']]
list(gp) #Grouped the respective data on basis of State
```

```python
gp = df.groupby(['State'])[['Confirmed','Deaths','Recovered']].sum().reset_index() #print the sum of the respective data in columns given
gp
```

```python
#Groupify on basis of Date and country/Region

#convert OD to date from object
df['OD'] = pd.to_datetime(df['OD'])
print(df.info())

dc = df.groupby(['Country/Region','OD'])[['Confirmed','Deaths','Recovered']].sum().reset_index()
dc
```

```python
#Scatter Graph
df.plot.scatter(x='Confirmed',y='Recovered') # x and y will not be string
```
