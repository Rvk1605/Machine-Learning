```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page = requests.get("https://www.flipkart.com/search?q=galaxy%20a12&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
print(page.content)
```

```python
soup  = BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
mobiles = soup.find_all("div",class_='_13oc-S')
print(len(mobiles))
# The find_all() function is a great function that helps to extract all HTML element that contains a given value.
# It takes an HTML element along with some attributes that tell it which class and id of div we need to select. 
# Here it is visible that I need a div element with a class attribute of maincounter-number. Now let’s search it.
```

```python
for mobile in mobiles:
    print(mobile.get_text())
```

```python
names = []
prices = []
frontCameras = []
for mobile in mobiles:
    name = mobile.find("div",class_='_4rR01T').text
    names.append(name)
    price = mobile.find("div",class_='_30jeq3 _1_WHN1').text
    prices.append(price)
    frontCamera = mobile.find("ul",class_='_1xgFaf')
    frontCameras.append(frontCamera.find_all('li')[2].text[24:])
```

```python
import pandas as pd
dic = {'names':names,'prices':prices,'frontCamera':frontCameras}
df = pd.DataFrame(dic)
df.head()
```

```python
df.to_csv('mobile_info.csv')
```
