```python
import matplotlib.pyplot as plt #pyplot is a submodule of matplotlib used to  make matplotlib work like MATLAB
import numpy as np

x= np.array([1,234])
y= np.array([1,20])

plt.plot(x,y) #plot the graph using x and y scale
plt.show()
```

# Default X Points

```python
y = np.array([1,3,3])
plt.plot(y) #By default x coordinates will be 0,1,2 corresponding to no of values in y
plt.show()
```

#Markers 

```python
x= np.array([1,234])
y= np.array([1,20])

plt.plot(x,y,marker="o") #marker is used to mark the points on the curve in graph
plt.show()
plt.plot(x,y,marker="D") 
plt.show()
plt.plot(x,y,"o")  #This will show only points
plt.show()
```

```python
plt.plot(x,y,marker = "o",linestyle="--",color = "g")  #This will show the curve in with diffrent styles and color
plt.show()
```

```python
x = np.array([1,22,33,35 ,42,57,67])
y = np.array([2,4,6,7,8,10,13])
plt.plot(x,y)
plt.title("Test Graph") #Graph Name
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
plt.show()
```

# Scatter Graph

```python
x = np.array([11,24,131,2,3,4,5,87,6,77,78,6])
y = np.array([13,45,48,4,5,6,7,8,9,3,4,6])

siz = np.array([20,50,100,200,500,1000,60,90,10,300,600,800])
plt.scatter(x,y,s=siz,alpha = 0.5)

plt.show()
```

# Bar Graph

```python
x = np.array(['A','B','C','D','E'])
y = np.array([2,3,4,5,6])
plt.bar(x,y,color='hotpink',width=0.3)
plt.show()

plt.barh(x,y) #for Horizontal
plt.show()
```

# Histogram

```python
#We see frequency Distribution in Histogram
# numpy.random.normal(loc = 0.0, scale = 1.0, size = None) 
# creates an array of specified shape and fills it with random values which is actually a part of Normal(Gaussian)Distribution.
# This is Distribution is also known as Bell Curve because of its characteristics shape.
# loc   : [float or array_like]Mean of 
# the distribution. 
# scale : [float or array_like]Standard 
# Derivation of the distribution. 
# size  : [int or int tuples].

x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

# PieChart

```python
y = np.array([25,35,21,15]) #Percentage Values in array
labels=["apple","orange","Grapes","Berry"]
exp = [0.2,0,0,0]
plt.pie(y,labels = labels,startangle=90,explode = exp,shadow = True) #Starts with 90 degree 
#Explode seperated the first ele with specified value

plt.legend(title = "four fruits")#Shows Label with meaning of every color
plt.show()
```
